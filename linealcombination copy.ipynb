# Análisis mediante álgebra lineal para graficar rendimientos esperados

### Introducción

¿Cómo mostrar gráficamente que una entidad mejora o empeora su rendimiento? Podemos dar una solución total (si conocemos todas las variables que dictan el rendimiento de una entidad) o bien parcial a esta pregunta (si no estamos seguros de conocer todas las variables que dictan un rendimiento) utilizando vectores, los cuales pueden representar una habilidad o hábito adquirido. La combinación lineal de estos vectores mostrarán una visualización general de las capacidades de la entidad. Si las habilidades o hábitos son positivos, estos harán que el vector resultante producto de la combinación lineal de los mismos se mostrará en aumento, en cambio si el hábito es negativo o bien el hecho de no poseer una habilidad hará que el vector resultante de la combinación lineal no crezca o bien retroceda.

### ¿Qué es un vector?

Desde un punto de vista geométrico, en el espacio bidimensional o tridimensional con el que estamos familiarizados intuitivamente, un vector puede ser visualizado como una flecha que apunta desde un punto de origen a otro punto. La dirección de la flecha indica la dirección del vector, y la longitud de la flecha indica su magnitud.
Formalmente, un vector en un espacio n-dimensional sobre el campo de los números reales R se puede representar como una tupla ordenada de n números reales:

v=(v1,v2,...,vn)
donde cada vi​ (para i=1,2,...,n) es un componente del vector en la i-ésima dimensión.

### ¿Qué es una combinación lineal?

Dado un conjunto de vectores v1,v2,...,vn en un espacio vectorial V sobre un campo F (por ejemplo, el campo de los números reales R o el campo de los números complejos C), y dado un conjunto de escalares a1,a2,...,an en F, una combinación lineal de los vectores v1,v2,...,vn con los escalares a1,a2,...,an​ como coeficientes es un vector v definido por la suma:

v = a1v1+a2v2+⋯+anvn 

En esta expresión, cada término aivi representa el producto del escalar ai​ por el vector vi, y el vector v es el resultado de sumar estos productos vectoriales. Los escalares ai pueden tomar cualquier valor en F.

### Elementos que representan la idea

* v => capacidad general o parcial de la entidad
* escalares, a1, a2, ... an => talentos, defectos, ventajas o habilidades naturales de la entidad.
* vectores, v1, v2, ... vn => talentos, defectos, ventajas o habilidades adquiridos por la entidad.
* espacio vectorial V => contexto en el cual se observan los vectores.
* campo F => magnitudes con las cuales se representarán los escalares.

### Aplicación práctica para medir rendimiento esperado para un jugador de ajedrez

Utilizando la idea anterior, podemos graficar el rendimiento **parcialmente esperado** (ya que tomaré a modo de ejemplo solo algunos factores) para un jugador de ajedrez basado en los ámbitos que la comunidad ajedrecística considera esenciales [agregar referencias]. Para este ejemplo abordaremos las magnitudes del 0 a 100, pero en otros casos estas magnitudes podrían sin problema representarse por números negativos, donde 0 es no tener conocimientos al respecto y 100 es tener un conocimiento acabado y experto en el tema (se debe tener cuidado con la interpretación del 100, dado que probablemente nunca un jugador de ajedrez llegue a abordar esta magnitud máxima dada la complejidad casi infinita de este juego).

##### Escalares
 1. Memoria => m
 2. Análisis => a
 3. Creatividad => c
##### Vectores
Para este ejemplo ***x*** será la capacidad del jugador e ***y*** será la cantidad de partidas que ha jugado
 1. Estudio de aperturas => (x1,y1)
 2. Generar planes de medio juego => (x2,y2)
 3. Estudio de finales => (x3,y3)
##### Espacio vectorial 
* rendimiento esperado => Aspectos esenciales que debe dominar un jugador de ajedrez
##### Formula general
* rendimiento esperado = m (x1,y1) + c (x2,y2) + a (x3,y3)

##### Aplicación 
Mediremos rendimiento esperado de un jugador principiante y un gran maestro de ajedrez

```python
import numpy as np
import matplotlib.pyplot as plt
```

 1. *jugador principiante* : supongamos que este jugador principiante sólo ha jugado 20 partidas de ajedrez (escalar 1 para indicar aspectos considerados 'normales')
*  1 (7,20) + 1 (3,20) + 1 (1,20) = rendimiento esperado
*  1 (7,20) + 1 (3,20) + 1 (1,20) = [22, 120]

```python
v1 = np.array([7, 20])
v3 = np.array([3, 20])
v4 = np.array([1, 20])
escalares = np.array([2, 2, 2])

combLinial = escalares[0] * v1 + escalares[1] * v3 + escalares[2] * v4
print(combLinial)

plt.figure(figsize=(10, 8))
plt.axvline(x=0, color='grey', lw=1)
plt.axhline(y=0, color='grey', lw=1)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', width=0.004, alpha=0.5)
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='b', width=0.004, alpha=0.5)
plt.quiver(0, 0, v4[0], v4[1], angles='xy', scale_units='xy', scale=1, color='y', width=0.004, alpha=0.5)

plt.quiver(0, 0, combLinial[0], combLinial[1], angles='xy', scale_units='xy', scale=1, color='black', width=0.005)

plt.xlim(0, combLinial[1]+10)
plt.ylim(0, combLinial[1]+10)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Combinación Lineal de 3 Vectores')
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

2. *gran maestro de ajedrez* : un GM probablemente ha jugado cientos de miles de partidas, pero para este **ejemplo** consideraremos 1000 juegos (escalar 3 para indicar aspectos considerados avanzados en un GM)
* 3 (90,1000) + 3 (90,1000) + 3 (90,1000) = rendimiento esperado
* 3 (90,1000) + 3 (90,1000) + 3 (90,1000) = [1200, 15000]

```python
v1 = np.array([90, 1000])
v3 = np.array([80, 1000])
v4 = np.array([70, 1000])
escalares = np.array([5, 5, 5])

combLinial = escalares[0] * v1 + escalares[1] * v3 + escalares[2] * v4
print(combLinial)

plt.figure(figsize=(10, 8))
plt.axvline(x=0, color='grey', lw=1)
plt.axhline(y=0, color='grey', lw=1)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', width=0.004, alpha=0.5)
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='b', width=0.004, alpha=0.5)
plt.quiver(0, 0, v4[0], v4[1], angles='xy', scale_units='xy', scale=1, color='y', width=0.004, alpha=0.5)

plt.quiver(0, 0, combLinial[0], combLinial[1], angles='xy', scale_units='xy', scale=1, color='black', width=0.005)

plt.xlim(0, combLinial[1]+10)
plt.ylim(0, combLinial[1]+10)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Combinación Lineal de 3 Vectores')
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

### Comparación con otras formas de representar rendimientos esperados

Otra forma comunmente utilizada para mostrar rendimientos o atributos es mediante graficos radiales. Los gráficos radiales muestran el valor que debería tener un atributo asociado a una entidad. El problema de visualizar los atributos de esta forma es que no se tiene un valor que sume totalmente los atributos e impide poder establecer un valor "general". Los graficos radiales presentan el problema de no poder amplificar sus valores por factores adicionales a la entidad, como se presentaba en la introducción si la entidad es una persona, no es posible representar claramente como una habilidad, cualidad o hábito impacta en el atributo de la entidad.

### Conclusiones

* La entidad puede ser prácticamente cualquier cosa, tales como, productos comercializables, jugador de un videojuego, deportista, etc. Mediante combinaciones lineales podemos obtener un valor final al sumar todos los atributos de la entidad, ya sean positivos o negativos.
* Para asignar pesos adecuados a los atributos que serán representados por los vectores y escalares es necesario como siempre contar con datos que respalden dichas magnitudes para obtener un valor lo más cercano a la realidad.
* No es posible obtener un valor adecuado si no conocemos el TOTAL de los atributos de la entidad.
